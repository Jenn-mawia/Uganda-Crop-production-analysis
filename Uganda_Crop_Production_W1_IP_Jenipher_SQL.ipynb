# Display list of Sub Counties, their population and areas


```
# this code displays the list of SubCounties, their population and areas
# 
%%sql
SELECT "SUBCOUNTY_NAME", POP, Area FROM Uganda;
```

# Sort the list of districts by total crop area

```
# sort the list of districts by total crop area in descending order
%%sql
SELECT "DISTRICT_NAME", SUM("Crop_Area_Ha") AS "Tot_Crop_Area_" FROM Uganda
GROUP BY "DISTRICT_NAME"
ORDER BY SUM("Crop_Area_Ha") DESC;
```

# Select only Sub Counties from Moroto district and order them alphabetically and show production of sorghum


```
%%sql
SELECT "SUBCOUNTY_NAME", "S_Prod_Tot" FROM Uganda
WHERE "DISTRICT_NAME" = "MOROTO"
ORDER BY "SUBCOUNTY_NAME" ASC;
```

# Compute total Maize production per district


```
%%sql
SELECT "DISTRICT_NAME", SUM("M_Prod_Tot") AS "Tot_M_Prod" FROM Uganda
GROUP BY "DISTRICT_NAME";
```

# Compute the number of sub counties where maize is produced and the total maize production per district

```
%%sql
SELECT "DISTRICT_NAME", SUM("M_Prod_Tot") AS "TOT_M_Pro_in_Dist", 
COUNT("SUBCOUNTY_NAME") AS "NO_OF_SubCounties_Prod_M" FROM Uganda
WHERE "M_Prod_Tot" != 0
GROUP BY "DISTRICT_NAME"; 
```

# Compute the overall crop area in all sub counties where population is over 20000


```
%%sql
SELECT "SUBCOUNTY_NAME", SUM("Crop_Area_Ha") AS "Overall_Crop_Area", POP FROM Uganda
WHERE POP > 20000
GROUP BY "SUBCOUNTY_NAME";
```

# Sort the maize production in descending order by districts, only taking into account Sub Counties where maize area is larger than sorghum area, and display number of sub counties per district matching this criteria

```

%%sql
SELECT "DISTRICT_NAME", SUM("M_Area_Ha") AS "TOT_M_Area", SUM("S_Area_Ha") 
AS "Tot_S_Area", SUM("M_Prod_Tot") AS "TOT_M_Pro_in_Dist",
COUNT("SUBCOUNTY_NAME") AS "NO_OF_SubCounties" FROM Uganda
WHERE "M_Area_Ha" >= "S_Area_Ha"
GROUP BY "DISTRICT_NAME"
ORDER BY "DISTRICT_NAME" DESC; 
```
